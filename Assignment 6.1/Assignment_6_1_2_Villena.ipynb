```python
import numpy as np
import pandas as pd

df = pd.read_csv('diabetes.csv')
```

```python
#Identify column names
df.columns
```

```python
#Identify the data types
df.dtypes
```

```python
#Display the total nunber of records

n = len(df)
print(f"The total number of records is {n}")
```

```python
#Print the first 20 records
print('The first 20 records are', df.head(20))
```

```python
#Print the last 20 records
print('The last 20 records are', df.tail(20))
```

```python
#Change the Outcome column to Diagnosis
df.rename(columns={'Outcome': 'Diagnosis'}, inplace=True)

df.columns
```

```python
#Create a new column that displays "diabetes and "without diabetes"
classification = {0: 'without Diabetes', 1: 'with Diabetes'}
df['Classification'] = df['Diagnosis'].map(classification)

df.head()
```

```python
#Create a dataframe withdiabetes that gathers data with diabetes
df_withdiabetes = df[df['Classification'] == 'with Diabetes']

print(len(df_withdiabetes))
df_withdiabetes.head()

```

```python
#Create a dataframe withdiabetes that gathers data no diabetes
df_nodiabetes = df[df['Classification'] == 'without Diabetes']

print(len(df_nodiabetes))
df_nodiabetes.head()
```

```python
#Create a dataframe that gathers data with age 0 to 19
df_pedia = df[(df['Age'] >= 0) & (df['Age'] <= 19)]

print(len(df_pedia))
df_pedia.head()
```

```python
#Create a dataframe that gathers Adult > 19
df_adult = df[df['Age'] >= 19]

print(len(df_adult))
df_adult.head()
```

```python
#Use numpy to compute for the average age and glucose value

print('The Average Age is:',np.mean(df['Age']))
print('The Average Glucose Level is:',np.mean(df['Glucose']))
```

```python
#Use numpy to compute for the median age and glucose value

print('The Median Age is:',np.median(df['Age']))
print('The Median Glucose Level is:',np.median(df['Glucose']))
```

```python
#Use numpy to get the middle values of age and glucose

median_age = np.median(df['Age'])
median_less1_age = median_age - 1
median_greater1_age = median_age + 1

middle_val_age = [median_less1_age, median_age, median_greater1_age]

median_gluco = np.median(df['Glucose'])
median_less1_gluco = median_gluco - 1
median_greater1_gluco = median_gluco + 1

middle_val_gluco = [median_less1_gluco, median_gluco, median_greater1_gluco]


print('Middle Values for Glucose:', middle_val_gluco)
print('Middle Values for Age:', middle_val_age)
```

```python
#Use numpy to get the standard deviation of skin thickness

print('The Standard Deviation of Skin Thickness is', np.std(df['SkinThickness']))
```
